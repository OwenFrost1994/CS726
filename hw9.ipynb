```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(2,4,201)
```

```python
y = np.power(x,2)+1
yf = np.linspace(0,0,201)
```

```python
fig,ax = plt.subplots(1,1,figsize=(8,8))
ax.plot(x, y, c = 'b')
ax.plot(x, yf, c = 'r')
ax.scatter(2, 5, c = 'g', s = 20)
ax.set_xlabel('$x$', fontsize=20)
ax.set_ylabel('$y$', fontsize=20)
ax.legend(['f(x)','feasible set', 'optimal point'], fontsize=20)
```

```python
x1 = np.linspace(0,5,201)
lamb = 1
y1 = np.power(x1,2)+1+ lamb*(x1-2)*(x1-4)
lamb = 2
y2 = np.power(x1,2)+1+ lamb*(x1-2)*(x1-4)
lamb = 3
y3 = np.power(x1,2)+1+ lamb*(x1-2)*(x1-4)
lamb = 5
y4 = np.power(x1,2)+1+ lamb*(x1-2)*(x1-4)
```

```python
fig,ax = plt.subplots(1,1,figsize=(8,8))
ax.plot(x, y, c = 'b')
ax.plot(x1, y1, '--', c = 'b')
ax.plot(x1, y2, '-.', c = 'b')
ax.plot(x1, y3, ':', c = 'b')
ax.plot(x1, y4, '<', c = 'b')
ax.plot(x1, 5*np.ones(201), c = 'r')
ax.scatter([x1[np.argmin(y1)],x1[np.argmin(y2)],x1[np.argmin(y3)],x1[np.argmin(y4)]], [min(y1),min(y2),min(y3),min(y4)], c = 'g', s = 100)
ax.set_xlabel('$x$', fontsize=20)
ax.set_ylabel('$y$', fontsize=20)
ax.legend(['f(x)','$\lambda=1$','$\lambda=2$','$\lambda=3$','$\lambda=5$','p*', '$inf_x L(x,lambda)$'], fontsize=20)
```

```python
x1 = np.linspace(0,5,201)
inff = np.zeros(100)
infx = np.zeros(100)
inflamb = np.zeros(100)
for i in range(100):
    lamb = 0.1 * i
    yi = np.power(x1,2)+1+ lamb*(x1-2)*(x1-4)
    infx[i] = x1[np.argmin(yi)]
    inff[i] = min(yi)
    inflamb[i] = lamb
```

```python
fig,ax = plt.subplots(1,1,figsize=(8,8))
ax.plot(x, y, c = 'b')
ax.plot(infx, inff, '-<', c = 'b')
ax.plot(x1, 5*np.ones(201), c = 'r')
ax.set_xlabel('$x$', fontsize=20)
ax.set_ylabel('$y$', fontsize=20)
ax.legend(['f(x)','g','p*'], fontsize=20)
```

```python
fig,ax = plt.subplots(1,1,figsize=(8,8))
ax.plot(x, y, c = 'b')
ax.plot(infx, inff, '-<', c = 'b')
ax.plot(x1, 5*np.ones(201), c = 'r')
ax.scatter(infx[np.argmax(inff)],max(inff),c='g',s=100)
ax.set_xlabel('$x$', fontsize=20)
ax.set_ylabel('$y$', fontsize=20)
ax.legend(['f(x)','g','p*','$lamb^{*}$ ='+str(inflamb[np.argmax(inff)])], fontsize=20)
```

```python
u = np.linspace(-5,5,41)
x1 = np.linspace(0,5,201)
inff = np.zeros(100)
infx = np.zeros(100)
inflamb = np.zeros(100)
ps = np.zeros(u.shape[0])
pu = np.zeros(u.shape[0])
for j in range(u.shape[0]):
    for i in range(100):
        lamb = 0.1 * i
        yi = np.power(x1,2)+1+ lamb*((x1-2)*(x1-4)-u[j])
        infx[i] = x1[np.argmin(yi)]
        inff[i] = min(yi)
        inflamb[i] = lamb
    ps[j] = max(inff)
    pu[j] = u[j]
```

```python
fig,ax = plt.subplots(1,1,figsize=(8,8))
ax.plot(pu, ps, '-', c = 'b')
ax.set_xlabel('$u$', fontsize=20)
ax.set_ylabel('$p^{*}$', fontsize=20)
```

```python
x = np.linspace(-4,4,201)
pa = np.zeros(201)
pc = np.zeros(201)
for i in range(201):
    if x[i]<0:
        pa[i] = -x[i]
        pc[i] = x[i]**2
    else:
        if x[i] >= 0 and x[i] <= 1:
            pa[i] = 0
            pc[i] = 0
        else:
            pa[i] = x[i]-1
            pc[i] = (x[i]-1)**2
```

```python
fig,ax = plt.subplots(1,1,figsize=(8,8))
ax.plot(x, pa, '-', c = 'b')
ax.plot(x, pc, '-', c = 'r')
ax.set_xlabel('$x$', fontsize=20)
ax.set_ylabel('$p$', fontsize=20)
ax.legend(['$p^{a}$', '$p^{c}$'], fontsize=20)
```
